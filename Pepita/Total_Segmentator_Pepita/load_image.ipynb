```python
import nibabel as nib

# Load your NIfTI file
img = nib.load("/home/abigail/mphys/mandible.nii.gz")

# Convert image data to a NumPy array
data = img.get_fdata()

import matplotlib.pyplot as plt

# Show the middle slice in the z direction
slice_index = data.shape[2] // 2
plt.imshow(data[:, :, slice_index], cmap="gray")
plt.axis("off")
plt.show()

# Loop through slices along z-axis
for i in range(0, data.shape[2], 10):  # every 10th slice
    plt.imshow(data[:, :, i], cmap="gray")
    plt.title(f"Slice {i}")
    plt.axis("off")
    plt.show()
    



```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import nibabel as nib
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output

# --- Load NIfTI image ---
img = nib.load("/home/abigail/mphys/cropped_CT_WHOLE_CNS_fixed.nii")
X = img.get_fdata()
X = (X - np.min(X)) / (np.max(X) - np.min(X))

# --- Display function ---
def show_slice(slice_index):
    plt.figure(figsize=(5, 5))
    plt.imshow(X[:, :, slice_index].T, cmap='gray', origin='lower')
    plt.title(f"Slice {slice_index}/{X.shape[2]-1}")
    plt.axis("off")
    clear_output(wait=True)
    display(plt.gcf())
    plt.close()

# --- Slider widget ---
slider = widgets.IntSlider(
    value=X.shape[2] // 2,
    min=0,
    max=X.shape[2] - 1,
    step=1,
    description='Slice:',
    continuous_update=True,
    layout=widgets.Layout(width='80%')
)

widgets.interact(show_slice, slice_index=slider)
```
