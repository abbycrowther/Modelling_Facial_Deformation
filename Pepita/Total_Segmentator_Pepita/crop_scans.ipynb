```python
import numpy as np

import nibabel as nib

nii_file = '/home/abigail/mphys/CT_WHOLE_CNS.nii' 
img = nib.load(nii_file)
data = img.get_fdata()

crop_in_x_start = 100
crop_in_x_end = -100
crop_in_y_start = 100
crop_in_y_end = -100
crop_in_z = -150 #negative because it only wants the last elements

cropped_data = data[crop_in_x_start:crop_in_x_end, crop_in_y_start:crop_in_y_end:, crop_in_z:]

cropped_img = nib.Nifti1Image(cropped_data, affine=img.affine, header=img.header)

nib.save(cropped_img, '/home/abigail/mphys/CT_WHOLE_CNS_cropped.nii')

print("Cropped shape:", cropped_data.shape)
```
