```python
import matplotlib.pyplot as plt
import numpy as np
import csv

transform_angles=[]
dice_coefficient=[]

with open('file.csv','r') as csvfile:
    lines = csv.reader(csvfile, delimiter=',')
    for row in lines:
        transform_angles.append(row[0])
        dice_coefficient.append(int(row[1]))

plt.scatter(transform_angles, dice_coefficient, color = 'g',s = 100)
plt.xticks(rotation = 25)
plt.xlabel('Transform Angle (Â°)')
plt.ylabel('Dice Coefficient')
plt.title('Dice Coefficient of Segmentations at Different Transform Angles', fontsize = 20)

plt.show()
```
